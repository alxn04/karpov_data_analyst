–í —Ñ–∞–π–ª–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ç–æ–º –∫–∞–∫–∏–µ –∫–æ–º–ø–∞–Ω–∏–∏ —Å –∫–µ–º –∏ –Ω–∞ –∫–∞–∫—É—é —Å—É–º–º—É –∑–∞–∫–ª—é—á–∞–ª–∏ —Å–¥–µ–ª–∫–∏, –∞ —Å–∞–º–æ–µ –≤–∞–∂–Ω–æ–µ ‚Äì –Ω–∞—Å–∫–æ–ª—å–∫–æ –æ–Ω–∏ –ø—Ä–∏ —ç—Ç–æ–º –æ–ø–æ–∑–¥–∞–ª–∏.<br>
–ü—Ä–æ—á–∏—Ç–∞–π—Ç–µ —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ. –î–∞–ª–µ–µ –ø–µ—Ä–µ–≤–µ–¥–∏—Ç–µ delay –≤ timedelta —Ñ–æ—Ä–º–∞—Ç.  

```python
delays = pd.read_csv("https://stepik.org/media/attachments/course/72204/delays.csv")
delays.head()
```

```python
delays.delay = pd.to_timedelta(delays.delay.str.replace('-',''))
```

```python
delays.head()
```

–†–∞–∑–±–µ–π—Ç–µ –∫–æ–ª–æ–Ω–∫—É delay –Ω–∞ 3 –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –∏ –ø–æ–º–µ—Å—Ç–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–æ–ª–æ–Ω–∫—É delay_categorical.

```python
delays['delay_categorical'] = pd.cut(delays.delay,3)
```

```python
delays.head()
```

–ü–µ—Ä–µ–∑–∞–¥–∞–π—Ç–µ –∫–æ–ª–æ–Ω–∫—É delay_categorical, —á—Ç–æ–±—ã –∑–Ω–∞—á–µ–Ω–∏—è –≤ –Ω–µ–π –±—ã–ª–∏: <br>
‚Ä¢	'less than 1 day' ‚Äì –≤—Ä–µ–º—è –æ—Ç 0 –¥–æ 1 –¥–Ω—è <br>
‚Ä¢	'1-2 days' ‚Äì –æ—Ç 1 –¥–æ 2 –¥–Ω–µ–π <br>
‚Ä¢	'2-3 days' ‚Äì –æ—Ç 2 –¥–æ 3 –¥–Ω–µ–π <br>
‚Ä¢	 'more than 3 days' ‚Äì –±–æ–ª—å—à–µ 3 –¥–Ω–µ–π <br>


```python
bins = [pd.to_timedelta(0,'D'), 
        pd.to_timedelta(1,'D'), # –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 0 –¥–æ 1
        pd.to_timedelta(2,'D'), # –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 1 –¥–æ 2
        pd.to_timedelta(3,'D'), # –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 2 –¥–æ 3
        pd.to_timedelta(5,'D')] # –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 3 –¥–æ 5
labels = ['less than 1 day','1-2 days','2-3 days','more than 3 days']
```

```python
pd.cut(delays.delay, bins)
```

```python
delays['delay_categorical'] = pd.cut(delays.delay,bins,labels = labels)
```

```python
delays.head()
```

–¢–µ–ø–µ—Ä—å –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–π –±–∞—Ä–ø–ª–æ—Ç —Å —Ç–µ–º, –Ω–∞—Å–∫–æ–ª—å–∫–æ —á–∞—Å—Ç–æ –∑–∞–¥–µ—Ä–∂–∏–≤–∞—é—Ç—Å—è —Å–¥–µ–ª–∫–∏.

```python
import plotly.express as px
```

```python
delays_counts = delays.delay_categorical.value_counts()
delays_counts
```

```python
px.bar(delays_counts)
```

```python
ads_data = pd.read_csv("data/6_ads/ads_data.csv", parse_dates = ['time'])
ads_data.head()
```

```python
# –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫–ª–∏–∫–æ–≤ –∏ –ø–æ–∫–∞–∑–æ–≤ –∫–∞–∂–¥–æ–≥–æ –æ–±—ä—è–≤–ª–µ–Ω–∏—è
ads_events = ads_data.groupby(['ad_id', 'event'], as_index = False).event.value_counts() \
        .pivot(index = 'ad_id', columns = 'event', values = 'count').fillna(0) \
        .reset_index()
ads_events
```

```python
# –°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∫–ª–∏–∫–æ–≤ –ø–æ –æ–±—ä—è–≤–ª–µ–Ω–∏—è–º
round(ads_events.click.mean())
```

```python
# –°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –ø–æ–∫–∞–∑–æ–≤ –æ–±—ä—è–≤–ª–µ–Ω–∏–π
round(ads_events.view.mean())
```

2. –ü—Ä–æ–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä—É–π—Ç–µ —á–∏—Å–ª–æ –ø–æ–∫–∞–∑–æ–≤, –∞ –∑–∞—Ç–µ–º –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–∞–∫–æ–π –≥—Ä–∞—Ñ–∏–∫, —á—Ç–æ–±—ã –º–æ–∂–Ω–æ –±—ã–ª–æ –ø–æ–Ω—è—Ç—å —Ñ–æ—Ä–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –í—ã–±–µ—Ä–∏—Ç–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π –≥—Ä–∞—Ñ–∏–∫:

```python
sns.displot(np.log(ads_events.view), kde = 'True', stat = 'density')
plt.title("–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–∫–∞–∑–æ–≤ –æ–±—ä—è–≤–ª–µ–Ω–∏–π")
plt.ylabel("")
plt.xlabel("–ß–∏—Å–ª–æ –ø–æ–∫–∞–∑–æ–≤")
```

3. –¢–µ–ø–µ—Ä—å –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫–∞–∑–æ–≤ –Ω–∞ 1 –æ–±—ä—è–≤–ª–µ–Ω–∏–µ (ad_id) –ø–æ –¥–Ω—è–º, –Ω–µ —É—á–∏—Ç—ã–≤–∞—è –æ–±—ä—è–≤–ª–µ–Ω–∏—è, —É –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ –±—ã–ª–æ –ø–æ–∫–∞–∑–æ–≤ (–æ–±—ã—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ). –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ä–µ–¥–Ω–µ–µ –ø–æ –ø–æ–ª—É—á–µ–Ω–Ω–æ–º—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É —Å –æ–∫–Ω–æ–º –≤ 2 —ç–ª–µ–º–µ–Ω—Ç–∞.

```python
ads_mean = ads_data[ads_data.event == 'view'].groupby(['date','ad_id'], as_index = False) \
        .event.value_counts().pivot_table(index = 'ad_id', columns = 'date', values = 'count') \
        .mean()
ads_mean
```

```python
rolling_ads_mean = ads_data[ads_data.event == 'view'].groupby(['date','ad_id'], as_index = False) \
        .event.value_counts().rename(columns={'count':'view_count'}) \
        .groupby('date').view_count.mean().rolling(2).mean().round(0)
rolling_ads_mean
```
