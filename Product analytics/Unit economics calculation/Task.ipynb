# Задание

Вы устроились аналитиком в интернет-магазин одежды. Продакт-менеджер хочет добиться роста в прибыли на 25% и попросил вас помочь ему со сценариями.

Данные за прошлый месяц:

* Выручка - 5,1 млн.рублей
* Себестоимость товаров - 2,8 млн. рублей (=> 45% маржинальность)
* Расходы на рекламу - 1,5 млн. рублей
* Заказы - 7060
* Покупатели - 6 441
* Посетители - 690 000


Посчитайте плечи метрик для того, чтобы определить, что поможет продакт-менеджеру добиться цели.

Финансисты посчитали, что: 
1.	+0,1% (речь про процентные пункты) к конверсии обойдется компании в 250К рублей
2.	Стоимость привлечения каждого дополнительного пользователя равна 4 рублям
3.	100 рублей к среднему чеку пользователя обойдутся в 200К рублей на маркетинг
4.	+0,1 к кол-ву повторных покупок на пользователя обойдутся в 400К рублей


Определите наиболее выгодный для компании вариант повышения прибыли и расскажите менеджерам, что им делать (как повысить прибыль). Сделайте презентацию в Power Point
